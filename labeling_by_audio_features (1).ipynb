```python
def smart_label(row):
    v = row['valence']
    e = row['energy']
    d = row['danceability']
    a = row['acousticness']
    t = row['tempo']
    s = row['speechiness']

    if v > 0.7 and e > 0.6 and d > 0.6:
        return "happy"
    elif v < 0.3 and e < 0.4 and a > 0.5:
        return "sad"
    elif e > 0.75 and v < 0.4:
        return "angry"
    elif e > 0.75 and v > 0.4 and t > 110:
        return "energetic"
    elif v > 0.6 and a > 0.5 and e < 0.5:
        return "romantic"
    elif v < 0.5 and e < 0.4 and t < 90 and a > 0.6:
        return "calm"
    elif v > 0.5 and e < 0.5 and t < 100:
        return "relaxed"
    else:
        return "calm"  # fallback


# Apply the new formula
df['mood'] = df.apply(smart_label, axis=1)

# Review
print(df['mood'].value_counts())
df[['track_name', 'valence', 'energy', 'danceability', 'acousticness', 'tempo', 'mood']].head()
```

```python
import pandas as pd
import requests
import time
from tqdm import tqdm

# Load your reduced Spotify dataset
df = pd.read_csv("spotify_2500_sample.csv")  # Make sure this file has 'track_name' and 'track_artist' columns

# Prepare list to hold lyrics
lyrics_list = []

# Loop through each row
for i, row in tqdm(df.iterrows(), total=len(df)):
    title = str(row['track_name'])
    artist = str(row['track_artist'])

    try:
        url = f"https://api.lyrics.ovh/v1/{artist}/{title}"
        response = requests.get(url)

        if response.status_code == 200:
            lyrics = response.json().get('lyrics', None)
            lyrics_list.append(lyrics)
        else:
            lyrics_list.append(None)

        time.sleep(0.5)  # be nice to the API

    except Exception as e:
        print(f"❌ Error fetching lyrics for: {title} by {artist} — {e}")
        lyrics_list.append(None)

# Add the lyrics column to the dataset
df['lyrics'] = lyrics_list


```

```python
import pandas as pd
import re
from tqdm import tqdm
from transformers import pipeline

# Step 1: Load your dataset
df = pd.read_csv("spotify_2500_with_lyrics.csv")  # Make sure it has a 'lyrics' column

# Step 2: Clean the lyrics
def clean_lyrics(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r'\[.*?\]', '', text)               # Remove [Chorus], [Verse], etc.
    text = re.sub(r'[^a-zA-Z\s]', '', text)           # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()          # Normalize whitespace
    return text

df['clean_lyrics'] = df['lyrics'].apply(clean_lyrics)

# Step 3: Load sentiment/emotion classifier (BERT-based)
emotion_model = pipeline(
    "text-classification",
    model="bhadresh-savani/distilbert-base-uncased-emotion",
    truncation=True
)

# Step 4: Predict sentiment for each row
sentiment_labels = []

for text in tqdm(df['clean_lyrics'], desc="Predicting sentiment"):
    if text and len(text) > 10:
        try:
            result = emotion_model(text[:512])[0]  # BERT limit is 512 tokens
            sentiment_labels.append(result['label'].lower())
        except:
            sentiment_labels.append("unknown")
    else:
        sentiment_labels.append("unknown")

df['lyrics_sentiment'] = sentiment_labels

# Step 5: Save the labeled dataset
df.to_csv("spotify_2500_with_sentiments.csv", index=False)
print("✅ Sentiment labels added and saved as 'spotify_2500_with_sentiments.csv'")
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("spotify_2500_with_sentiments.csv")

# Normalize tempo to [0, 1]
def normalize_tempo(t):
    return max(0, min((t - 40) / (200 - 40), 1))

# Compute arousal: energy + tempo + (1 - acousticness)
def compute_arousal(row):
    tempo_norm = normalize_tempo(row['tempo'])
    arousal = (row['energy'] + tempo_norm + (1 - row['acousticness'])) / 3
    return arousal

# Define nuanced mood logic
def refined_mood(row):
    val = row['valence']
    ar = compute_arousal(row)
    sp = row['speechiness']

    if val > 0.7 and ar > 0.6:
        return "energetic"
    elif val > 0.7 and ar <= 0.6:
        return "relaxed"
    elif val < 0.3 and ar < 0.4:
        return "sad"
    elif val < 0.3 and ar > 0.6:
        return "hype" if sp > 0.33 else "intense"
    elif val > 0.6 and 0.4 < ar < 0.7:
        return "joyful"
    elif val < 0.4 and 0.4 < ar < 0.7:
        return "melancholic"
    else:
        return "calm"

# Apply to dataset
df['mood'] = df.apply(refined_mood, axis=1)

# Save result
df.to_csv("spotify_2500_refined_moods.csv", index=False)
print("✅ Updated mood labels with nuanced categories.")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("spotify_2500_valence_arousal_moods.csv")

# 🧼 Clean mood labels (strip spaces, lowercase)
df['mood'] = df['mood'].astype(str).str.strip().str.lower()

# 🔍 See class distribution
print("Mood distribution:\n", df['mood'].value_counts())

# 🎯 Audio features to use
audio_features = ['valence', 'energy', 'danceability', 'tempo']
X = df[audio_features]
y = df['mood']

# ✂️ Split into 50% train, 50% test (with stratification)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.5,
    random_state=42,
    stratify=y  # preserve class proportions
)

# 🔢 Encode labels
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.transform(y_test)

# 🌳 Train Random Forest
model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model.fit(X_train, y_train_enc)

# 🧪 Predict on the test set
y_pred = model.predict(X_test)

# 📊 Evaluate
print("\nClassification Report on 50/50 Train-Test Split:")
print(classification_report(y_test_enc, y_pred, target_names=label_encoder.classes_))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Load dataset
df = pd.read_csv("spotify_2500_refined_moods.csv")

# Clean up mood and sentiment labels
df['mood'] = df['mood'].astype(str).str.strip().str.lower()
df['lyrics_sentiment'] = df['lyrics_sentiment'].astype(str).str.strip().str.lower()

# Define all audio features from Spotify
audio_features = [
    'valence', 'energy', 'danceability', 'tempo', 'acousticness',
    'loudness', 'speechiness', 'instrumentalness', 'liveness', 'mode', 'key', 'duration_ms'
]

# Drop rows with missing values in audio features or mood
df = df.dropna(subset=audio_features + ['mood'])

# 50/50 train-test split with stratification based on lyrics_sentiment (you can change this to mood too)
train_df, test_df = train_test_split(
    df,
    test_size=0.5,
    random_state=42,
    stratify=df['lyrics_sentiment']
)

# Prepare input/output for training
X_train = train_df[audio_features]
X_test = test_df[audio_features]

y_train = train_df['mood']
y_test = test_df['mood']

# Encode mood labels
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.transform(y_test)

# Train Random Forest on full audio features
rf_model_2 = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_model_2.fit(X_train, y_train_enc)

# Predict and evaluate
y_pred = rf_model_2.predict(X_test)

print("\n📊 Classification Report (50/50 split using ALL audio features):")
print(classification_report(y_test_enc, y_pred, target_names=label_encoder.classes_))
```
